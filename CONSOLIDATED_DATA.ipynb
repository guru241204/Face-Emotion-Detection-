```python
import cv2
import os
import pickle
import numpy as np
```

```python
data_dir = os.path.join(os.getcwd(),'data')
img_dir = os.path.join(os.getcwd(),'Images')
```

```python
print(os.getcwd()) #getcwd - get current working directory
```

```python
image_data = []
labels = []
```

```python
for i in os.listdir(img_dir):
    print(i)
```

```python
for i in os.listdir(img_dir):
    image = cv2.imread(os.path.join(img_dir,i))
    image = cv2.resize(image,(100,100))
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    image_data.append(image)
    labels.append(str(i).split("_")[0])
```

```python
image_data = np.array(image_data)    
labels = np.array(labels) 
```

```python
print(image_data)
```

```python
print(labels)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image_data[50],cmap="gray")
plt.show()
```

```python
with open(os.path.join(data_dir,"images.p"),'wb') as f:
    pickle.dump(image_data,f)
```

```python
with open(os.path.join(data_dir,"labels.p"),'wb') as f:
    pickle.dump(labels,f)
```
